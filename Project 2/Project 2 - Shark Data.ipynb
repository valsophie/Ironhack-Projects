### 4.  Change Age to INT 

```python
#type is stored as string, but I want to save it as a number
#it has some Nans -- so transformation to int not possible 
```

```python
# change everything that is not a digit to NAN

df["Age"] = np.where(df["Age"].str.isdigit() == False, np.NaN, df["Age"])
```

```python
#Fill all Nans with 0 in order to make it int 
df["Age"] = df["Age"].fillna(0).astype(int)
```

```python
#Fill all 0 back zu NAN
df["Age"] = np.where(df["Age"] == 0, np.NaN, df["Age"])
```
