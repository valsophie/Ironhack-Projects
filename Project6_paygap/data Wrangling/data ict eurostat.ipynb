```python
import pandas as pd
pd.options.display.max_rows= 200

#data from: https://ec.europa.eu/eurostat/web/products-datasets/-/isoc_sks_itsps
```

```python
df = pd.read_csv("data/ict_raw.csv")
```

```python
df.head(50)
```

```python
df = df[df.UNIT == "Percentage"]
df = df[df.SEX == "Females"]
```

```python
df.head()
```

```python
df.reset_index(drop=True)
```

```python
df.columns = map(str.lower, df.columns)
```

```python
df = df.rename(columns={"value": "prct_f_employed_ICT"})
```

```python
df = df.rename(columns={"geo": "location"})
```

```python
df = df.drop(["flag and footnotes", "unit", "sex"],axis=1)
```

```python
import pycountry
```

```python
type(list(pycountry.countries))
#list(pycountry.countries)[0]
```

```python
countries = {}
```

```python
countries["Germany (until 1990 former territory of the FRG)"] = "DEU"
```

```python
for country in list(pycountry.countries):
    countries[country.name] = country.alpha3
```

```python
list(pycountry.countries)[-1].name
```

```python
df = df.replace(countries)

df.reset_index(drop=True)
df.head(50)
```

```python
df["prct_f_employed_ICT"] = df["prct_f_employed_ICT"].str.replace(",",".")
```

```python
import numpy as np
df["prct_f_employed_ICT"] = np.where(df["prct_f_employed_ICT"] == ":", np.nan, df["prct_f_employed_ICT"])
```

```python
df['prct_f_employed_ICT'] = df['prct_f_employed_ICT'].astype(float)
```

```python
df.to_pickle("data/ict_clean.pkl")
```
