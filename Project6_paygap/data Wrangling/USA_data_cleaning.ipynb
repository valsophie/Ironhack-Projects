```python
#Total number of workers
df["Total Number of workers"]= df["Total Number of workers"].replace("–","NaN")
df["Total Number of workers"]= df["Total Number of workers"].str.replace(",","")
df["Total Number of workers"]= df["Total Number of workers"].astype(float)
```

```python
df
```

```python
#Median weekly earnings
df["Median weekly earnings"]= df["Median weekly earnings"].str.replace("–","NaN")
df["Median weekly earnings"]= df["Median weekly earnings"].str.replace(",","")
df["Median weekly earnings"]= df["Median weekly earnings"].astype(float)
#Number of workers Men
df["Number of workers Men"]= df["Number of workers Men"].str.replace("–","NaN")
df["Number of workers Men"]= df["Number of workers Men"].str.replace(",","")
df["Number of workers Men"]= df["Number of workers Men"].astype(float)
#Median weekly earnings Men
df["Median weekly earnings Men"]= df["Median weekly earnings Men"].replace("–","NaN")
df["Median weekly earnings Men"]= df["Median weekly earnings Men"].str.replace(",","")
df["Median weekly earnings Men"]= df["Median weekly earnings Men"].astype(float)
#Number of workers Women
df["Number of workers Women"]= df["Number of workers Women"].str.replace("–","NaN")
df["Number of workers Women"]= df["Number of workers Women"].str.replace(",","")
df["Number of workers Women"]= df["Number of workers Women"].astype(float)
#Median weekly earnings Women
df["Median weekly earnings Women"]= df["Median weekly earnings Women"].str.replace("–","NaN")
df["Median weekly earnings Women"]= df["Median weekly earnings Women"].str.replace(",","")
df["Median weekly earnings Women"]= df["Median weekly earnings Women"].astype(float)
#dropnans
df = df.dropna()
```
